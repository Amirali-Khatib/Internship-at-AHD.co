convert txt file separated json object to use the in dictionary which we want to convert it to data frame.

```python
def dict_converter(file_name):
  # initialization

  article_dict = dict()
  article_dict['content'] = list()
  article_dict['title'] = list()
  article_dict['lead'] = list()
  article_dict['agency'] = list()
  article_dict['date_created'] = list()
  article_dict['url'] = list()
  article_dict['image'] = list()
  article_dict['category'] = list()

  # main-body-of-function

  for article_counter in range(10000):
    line_counter = 0
    article = str()
    while line_counter < 16:
      line_counter += 1
      line = file_name.readline()
      article += line
    json_obj = json.loads(article)

    article_dict['title'].append( json_obj['_source']['title'])
    article_dict['content'].append( json_obj['_source']['content'])
    article_dict['agency'].append( json_obj['_source']['agency'])
    article_dict['category'].append( json_obj['_source']['category'])
    article_dict['date_created'].append( json_obj['_source']['date_created'])
    article_dict['image'].append( json_obj['_source']['image'])
    article_dict['lead'].append( json_obj['_source']['lead'])
    article_dict['url'].append( json_obj['_source']['url'])

  return article_dict
```

```python
output_dict = dict_converter(Article)
data_frame = pd.DataFrame(output_dict)
```

```python
df['content_norm'] = [normalizer.normalize(t) for t in df.content]
df['content_norm'] = [t.replace('\u200c', ' ') for t in df.content_norm]
df['content_w_tok'] = [word_tokenize(t) for t in df.content_norm]
df['content_s_tok'] = [sent_tokenize(t) for t in df.content_norm]
```

```python
WORDS = list()
```

```python
for r in range(len(df['content_w_tok'])):
  for w in df.loc[r]['content_w_tok']:
     WORDS.append(w)
```

```python
word_count = dict(Counter(WORDS))
```

```python
len(word_count)
```

```python
report = pd.DataFrame({'word':word_count.keys(), 'count':word_count.values()})
report.sort_values('count', ascending=False).head(10)
```

```python
def make_bigrams(x):
  bigrams = [b for l in x for b in zip(l.split(' ')[:-1], l.split(' ')[1:])]
  return bigrams
```

```python
df['content_bigrams'] = [make_bigrams(t) for t in df.content_s_tok]
```

```python
#df
```

```python
BIGRAMS = []
for r in range(len(df['content_bigrams'])):
  for b in df.loc[r]['content_bigrams']:
     BIGRAMS.append(b)
```

```python
len(BIGRAMS)
```

```python
report2_dict = dict(Counter(BIGRAMS))
report2_dict
```

```python
report_10 = dict(Counter(BIGRAMS).most_common(10))
report2_df = pd.DataFrame({'bigram': list(report_10.keys()), 'count': list(report_10.values())})
report2_df
```

We want to omit the persian stopwords from each article and analyse the same outputs which are processed above(without considering stopwords).
In first step we opened a file that contains persian stop words, and preprocessed the list of stopwords to make them usable for our main purpose.  

```python
stopwords_fa = open('/content/drive/MyDrive/persian_stopwords.txt','r')
stopwords_fa = stopwords_fa.readlines()
```

```python
def stopwords_editor(x):
  output = x.replace('\u200c', ' ')
  output = output.strip()
  return output
```

```python
stopwords_fa = [stopwords_editor(t) for t in stopwords_fa]
stopwords_fa
```

Omit stopwords from word list of all articles

```python
# for world count dictionary
for i in range(len(stopwords_fa)):
  if stopwords_fa[i] in list(word_count.keys()):
    del word_count[stopwords_fa[i]]
```

```python
report = pd.DataFrame({'word':word_count.keys(),'count':word_count.values()})
report.sort_values('count', ascending=False).head(10)
```

Omit bigrams which contained stopwords

```python
# for bigrams count dictionary
for b in list(report2_dict.keys()):
  if (b[0] in stopwords_fa) | (b[1] in stopwords_fa):
    del report2_dict[b] 
```

```python
report2_df = pd.DataFrame({'bigram': list(report2_dict.keys()),
                           'count': list(report2_dict.values())})
report2_df.sort_values('count', ascending=False).head(10)
```
